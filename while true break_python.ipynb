```python
## backuplog for user login and activity.
def saveactivity_log(user,action,action_detail):
    action_time = datetime.now()
    with open(r"D:/DSDS4002/Project_final/output/activitylog_%s_%s.txt" % (action_time.year, action_time.month), "a") as txtfile:
        if txtfile.tell() == 0:  # ตรวจสอบว่าไฟล์ว่างหรือไม่
            txtfile.write("User,Activity time,Action,Action Detail\n")
                   
        txtfile.write(f"{user},{action_time},{action},{action_detail}\n")

#for sql query and table display
def read_sql(sql_commands, inventory_db):
    df = pd.read_sql(sql_commands, inventory_db)
    display(df)
    return df

#for sql query w/o table display
def read_sql2(sql_commands, inventory_db):
    df = pd.read_sql(sql_commands, inventory_db)
    return df

#for sql insert+update+delete
def execute_sql(sql_commands, values):
    my_cursor.execute(sql_commands, values)
    inventory_db.commit()
        
####################################################################   User login   #####################################################################
print('=====================================================================================================================================================')
print('\033[1m========================================================= Welcome to inventory Database system ======================================================\033[0m')
print('=====================================================================================================================================================\n\n')

print('Please enter user name and password for login: ')
user = input('User name: ')
pwd = getpass('Password: ')

#####################################################################   DB connection   ##########################################################################
inventory_db = None
try:
    inventory_db = mysql.connector.connect( host='localhost',
                                     user=user,
                                     password=pwd,
                                     database='inventory')
    print('Successfully connected to the database.')
    
    my_cursor = inventory_db.cursor()

#########################################################################  Main program   ########################################################################
    action = 'Log in'
    action_detail = 'user log in'
    saveactivity_log(user,action,action_detail)
    choice = ''

    while choice not in ['q', 'Q']:
            print('====================================================================================================================================================')
            print('\033[1m============================================================== Main Menu ===========================================================================\033[0m')
            print('====================================================================================================================================================\n\n')
            print('Select option to explore')
            print('1. Query stock')
            print('2. Stock management')
            print('3. Report Summary')
            print('Q. quit')

            choice = input('Enter selection: ')
            
 #########################################################  Option 1.Product and Stock search ########################################################

            if choice == '1':
                search = ''
                print('=====================================================================================================================================================')
                print('\033[1m================================================================ Search Stock =======================================================================\033[0m')
                print('=====================================================================================================================================================\n\n')
                action = 'Query stock'
                while search not in ['q', 'Q']:

                    print('Select option to explore')
                    print('1.1 Find all stock balance search by date (1)')
                    print('1.2 Find all stock balance search by city and date (2)')
                    print('Q. Back to main Menu')
                    
                    search = input('Enter selection: ')
                    
                    if search == '1':
                        action_detail = 'Find all stock balance search by date'
                        search_int1 = input('please enter date (YYYY-MM-DD): ')
                        sql_commands_1 = '''select item_name, size,onhand as onhand_qty,
                                            COALESCE(purchase_quantity,0) as purchase_qty,COALESCE(sales_qty,0) as sales_qty,
                                            COALESCE(balance_stock,0) as balance_stock,inventory_value$
                                        from 
                                            (select b.inventory_id,b.onhand,b.cost_per_unit,b.amount,
                                            t1.purchase_quantity, t1.dollars,
                                            t2.sales_qty, item_name, size,
                                            b.onhand+COALESCE(t1.purchase_quantity,0)-COALESCE(t2.sales_qty,0) as balance_stock,
                                            b.onhand+COALESCE(t1.purchase_quantity, 0) as onhand_purchase,
                                            b.amount+t1.dollars as total_cost_onhand_purchase,
                                            round((b.amount+COALESCE(t1.dollars,0))/(b.onhand+COALESCE(t1.purchase_quantity, 0)),2) as avg_cost,
                                            round(((b.onhand+COALESCE(t1.purchase_quantity,0)-COALESCE(t2.sales_qty,0)) * (b.amount+COALESCE(t1.dollars,0))/(onhand+COALESCE(purchase_quantity, 0))),2) as inventory_value$
                                        from begin as b
                                        left join (select inventory_id, COALESCE(sum(dollars),0) as dollars, COALESCE(sum(quantity),0) as purchase_quantity 
                                                   from purchase where receiving_date between '2016-01-01' and '%s' group by inventory_id) as t1
                                        on b.inventory_id = t1.inventory_id
                                        left join (select inventory_id, COALESCE(sum(sales_qty),0) as sales_qty
                                                   from sale where sales_date between '2016-01-01' and '%s' group by inventory_id) as t2
                                        on b.inventory_id = t2.inventory_id
                                        left join inventory 
                                        on b.inventory_id = inventory.inventory_id ) t3
                                        order by 1;''' %(search_int1,search_int1)
                        
   
                        # Read from the database
                        df1 = read_sql(sql_commands_1, inventory_db)
                        export = input('Do you need to export stock balance? (Y/N): ')
                        if export == 'Y' or export == 'y':
                            df1.to_csv('D:/DSDS4002/Project_final/output/all_stock_balance.csv',index=False,mode='w')
                            print('Output data are under D:/DSDS4002/Project_final/output/\n')
                            print('file name:all_stock_balance.csv')

                        saveactivity_log(user,action,action_detail)
                        
                    elif search == '2':
                        action_detail = 'Find all stock balance search by city and date'
                        sql_commands_check = '''select distinct city from inventory'''
                        read_sql(sql_commands_check, inventory_db)
                        
                        search_int1 = input('please enter city code : ').upper()
                        search_int2 = input('please enter date (YYYY-MM-DD): ')
                        sql_commands_2 = '''select city, item_name, size,onhand as onhand_qty,
                                            COALESCE(purchase_quantity,0) as purchase_qty,COALESCE(sales_qty,0) as sales_qty,
                                            COALESCE(balance_stock,0) as balance_stock,inventory_value$
                                        from 
                                            (select city,b.inventory_id,b.onhand,b.cost_per_unit,b.amount,
                                            t1.purchase_quantity, t1.dollars,
                                            t2.sales_qty, item_name, size,
                                            b.onhand+COALESCE(t1.purchase_quantity,0)-COALESCE(t2.sales_qty,0) as balance_stock,
                                            b.onhand+COALESCE(t1.purchase_quantity, 0) as onhand_purchase,
                                            b.amount+t1.dollars as total_cost_onhand_purchase,
                                            round((b.amount+COALESCE(t1.dollars,0))/(b.onhand+COALESCE(t1.purchase_quantity, 0)),2) as avg_cost,
                                            round(((b.onhand+COALESCE(t1.purchase_quantity,0)-COALESCE(t2.sales_qty,0)) * (b.amount+COALESCE(t1.dollars,0))/(onhand+COALESCE(purchase_quantity, 0))),2) as inventory_value$
                                        from begin as b
                                        left join (select inventory_id, COALESCE(sum(dollars),0) as dollars, COALESCE(sum(quantity),0) as purchase_quantity 
                                                   from purchase where receiving_date between '2016-01-01' and '%s' group by inventory_id) as t1
                                        on b.inventory_id = t1.inventory_id
                                        left join (select inventory_id, COALESCE(sum(sales_qty),0) as sales_qty
                                                   from sale where sales_date between '2016-01-01' and '%s' group by inventory_id) as t2
                                        on b.inventory_id = t2.inventory_id
                                        left join inventory 
                                        on b.inventory_id = inventory.inventory_id ) t3
                                        where city like '%%%s%%'
                                        order by balance_stock desc;'''  %(search_int2,search_int2,search_int1)
                        # Read from the database
                        df2 = read_sql(sql_commands_2, inventory_db)
                        export = input('Do you need to export stock balance? (Y/N): ')
                        if export == 'Y' or export == 'y':
                            df2.to_csv('D:/DSDS4002/Project_final/output/city_stock_balance.csv',index=False,mode='w')
                            print('Output data are under D:/DSDS4002/Project_final/output/\n')
                            print('file name:city_stock_balance.csv')

                            saveactivity_log(user,action,action_detail)
                        
###############################################################  option 2.Update stock ###############################################################
###############################################################  option 2.1.Add stock  ###############################################################
                    
            elif choice == '2':
                update = ''
                action = 'Update stock'
                while update not in ['q', 'Q']:
                    print('====================================================================================================================================================')
                    print('\033[1m========================================================= Stock management =========================================================================\033[0m')
                    print('====================================================================================================================================================\n\n')
                    print('Select option to explore')
                    print('2.1 Add Stock (1)')
                    print('2.2 Update Stock (2)')
                    print('2.3 Delete Stock (3)')
                    print('Q. Back to main Menu')
                    
                    update = input('Enter selection: ')
                    if update == '1':
                        update_int1 = input('Select table to add :1)purchase 2)sale: ')

################################################################  Add Purchase  ################################################################

                        if update_int1 == 'purchase' or update_int1 =='1' :
                            action_detail = 'Add Purchase Stock'
                            while True:
                                purchase_1 = input('Enter PO number: ')
                                if purchase_1.isdigit():
                                    break
                                else:
                                    print("Invalid input. Please enter a numeric for PO number.")
                            
                            sql_commands_check = '''select * from purchase where po_number = '%s';''' %purchase_1
                            read_sql(sql_commands_check, inventory_db)                            
                            
                            purchase_2 = input('Enter inventory id: ')
                            purchase_3 = input('Enter PO date (YYYY-MM-DD): ')
                            
                            while True:
                                purchase_4 = input('Enter purchase Quantity: ')
                                purchase_5 = input('Enter purchase price: ')
                                if purchase_4.isdigit() and purchase_5.isdigit():
                                    break
                                else:
                                    print("Invalid input. Please enter a numeric for purchase Quantity and purchase price.")
                            
                            purchase_6 = float(purchase_4)*float(purchase_5)
                        
                            confirm =  input('Do you want to add into Purchase table (Y/N): ')
                            if confirm == 'Y' or confirm == 'y':
                                sql_commands_4 = """INSERT INTO inventory.purchase (po_number,inventory_id,po_date,quantity,purchase_price,dollars)
                                            VALUES (%s,%s,%s,%s,%s,%s) """
                                values = (purchase_1, purchase_2, purchase_3, purchase_4, purchase_5,purchase_6)
                        
                                # Execute SQL to add purchase
                                execute_sql(sql_commands_4, values)
                                print('Successfully inserted data to the database.')    
                                saveactivity_log(user,action,action_detail)
                                read_sql(sql_commands_check, inventory_db)
                            
#################################################################  Add Sale  ###################################################################                                
                            
                        elif update_int1 == 'sale' or update_int1 =='2' :
                            action_detail = 'Add Sale Stock'
                            while True:
                                sale_1 = input('Enter SO number: ')
                                if sale_1.isdigit():
                                    break
                                else:
                                    print("Invalid input. Please enter a numeric for SO number.")
                            
                            sql_commands_check = '''select * from sale where so_number = '%s';''' %sale_1
                            read_sql(sql_commands_check, inventory_db)
                            
                            sale_2 = input('Enter inventory id: ')
                            sale_3 = float(sale_2.split('_')[0])
                            while True:
                                sale_4 = input('Enter sale Quantity: ')
                                sale_5 = input('Enter sale price: ')
                                if sale_4.isdigit() and sale_5.isdigit():
                                    break
                                else:
                                    print("Invalid input. Please enter a numeric for sale Quantity and sale price.")
                            
                            sale_6 = float(sale_4)*float(sale_3)
                            sale_7 = input('Enter sale date (YYYY-MM-DD): ')
                            
                            confirm =  input('Do you want to add for sale table (Y/N): ')
                            if confirm == 'Y' or confirm == 'y':
                                sql_commands_5 = """INSERT INTO inventory.sale (so_number,inventory_id,city_id,sales_qty,sales_price,sales_dollar,sales_date)
                                            VALUES (%s,%s,%s,%s,%s,%s,%s) """
                                values = (sale_1, sale_2, sale_3, sale_4, sale_5,sale_6,sale_7)
                        
                                # Execute SQL to add sale
                                execute_sql(sql_commands_5, values)
                                print('Successfully inserted data to the database.')   
                                saveactivity_log(user,action,action_detail)
                                read_sql(sql_commands_check, inventory_db)

#############################################################  option 2.2.Update stock  ##############################################################
                    elif update == '2':
                        update_int1 = input('Select table to update :1)purchase 2)sale: ')   
            
######################################################################  Update Purchase  ########################################################################

                        if update_int1 == 'purchase' or update_int1 =='1' :
                            action_detail = 'Update Purchase Stock'
                            while True:
                                purchase_1 = input('Enter PO number: ')
                                sql_commands_check = '''select * from purchase where po_number = '%s';''' %purchase_1
                                df_check = read_sql(sql_commands_check, inventory_db)
                                if purchase_1.isdigit() and len(df_check) > 0:
                                    break
                                else:
                                    print("Invalid input. Please enter a numeric for PO number.")  
                            
                            purchase_2 = input('Enter inventory id: ')
                            purchase_3 = input('Enter PO date (YYYY-MM-DD): ')
                            purchase_7 = input('Enter Recieve date (YYYY-MM-DD): ')
                            purchase_7 = datetime.strptime(purchase_7, '%Y-%m-%d')
                            purchase_8 = input('Enter Invoice date (YYYY-MM-DD): ')
                            purchase_8 = datetime.strptime(purchase_8, '%Y-%m-%d')
                            purchase_9 = purchase_8 + timedelta(days=30)
                            purchase_9 = purchase_9.strftime('%Y-%m-%d')
                            
                            while True:
                                purchase_4 = input('Enter purchase Quantity: ')
                                purchase_5 = input('Enter purchase price: ')
                                if purchase_4.isdigit() and purchase_5.isdigit():
                                    break
                                else:
                                    print("Invalid input. Please enter a numeric for purchase Quantity and purchase price.")
                            
                            purchase_6 = float(purchase_4)*float(purchase_5)

                            confirm =  input('Do you want to update for purchase table (Y/N): ')
                            if confirm == 'Y' or confirm == 'y':                            

                                sql_commands_7 = """UPDATE inventory.purchase
                                               SET po_date = %s, receiving_date = %s, invoice_date = %s, pay_date = %s,
                                               purchase_price = %s, quantity = %s, dollars = %s
                                               WHERE po_number = %s and inventory_id = %s; """
                                values = (purchase_3, purchase_7,purchase_8,purchase_9,purchase_5,purchase_4,purchase_6,purchase_1,purchase_2)
                        
                                # Execute SQL to update purchase
                                execute_sql(sql_commands_7, values)
                                print('Successfully update data to the database.')
                                saveactivity_log(user,action,action_detail)
                            
                                sql_commands_check = '''select * from purchase where po_number = '%s' and inventory_id = '%s';''' %(purchase_1,purchase_2)
                                read_sql(sql_commands_check, inventory_db)

#########################################################################  Update Sale  ############################################################################
                                
                        elif update_int1 == 'sale' or update_int1 =='2' :
                            action_detail = 'Update Sale Stock'
                            while True:
                                sale_1 = input('Enter SO number: ')
                                sql_commands_check = '''select * from sale where so_number = '%s';''' %sale_1
                                df_check = read_sql(sql_commands_check, inventory_db)
                                if sale_1.isdigit() and len(df_check) > 0:
                                    break
                                else:
                                    print("Invalid input. Please enter a numeric for SO number.")                  
                            
                            sale_2 = input('Enter inventory id: ')
                            sale_3 = float(sale_2.split('_')[0])
                            while True:
                                sale_4 = input('Enter sale Quantity: ')
                                sale_5 = input('Enter sale price: ')
                                if sale_4.isdigit() and sale_5.isdigit():
                                    break
                                else:
                                    print("Invalid input. Please enter a numeric sale Quantity and sale price.")

                            sale_6 = float(sale_4)*float(sale_5)
                            sale_7 = input('Enter sale date (YYYY-MM-DD): ')
                            
                            confirm =  input('Do you want to update for sale table (Y/N): ')
                            if confirm == 'Y' or confirm == 'y':  
                            
                                sql_commands_8 = """UPDATE inventory.sale
                                               SET city_id = %s, sales_qty = %s, sales_price = %s, sales_dollar = %s, sales_date = %s
                                               WHERE so_number = %s and inventory_id = %s; """
                                values = (sale_3, sale_4, sale_5,sale_6,sale_7,sale_1,sale_2)
                            
                                # Execute SQL to update sale
                                execute_sql(sql_commands_8, values)
                                saveactivity_log(user,action,action_detail)
                                read_sql(sql_commands_check, inventory_db)                           
  

                    elif update == '3':
                        update_int1 = input('Select table to delete :1)purchase 2)sale: ')
            
#######################################################################  Delete Purchase  #########################################################################                                
                                
                        if update_int1 == 'purchase' or update_int1 =='1' :
                            action_detail = 'Delete Purchase Stock'
                            while True:
                                purchase_1 = input('Enter PO number: ')
                                sql_commands_check = '''select * from purchase where po_number = '%s';''' %purchase_1
                                df_check = read_sql(sql_commands_check, inventory_db)
                                if purchase_1.isdigit() and len(df_check) > 0:
                                    break
                                else:
                                    print("Invalid input. Please enter a numeric for PO number.")  
                            
                            purchase_2 = input('Enter inventory id: ')
                            print('======================================================================================================================================================')
                            print('\033[1m====================================================== product code selection to delete  =============================================================\033[0m')
                            print('======================================================================================================================================================\n\n')
                            sql_commands_check = '''select * from purchase where po_number = '%s' and inventory_id = '%s';''' %(purchase_1,purchase_2)
                            read_sql(sql_commands_check, inventory_db)

                            confirm =  input('Do you want to delete for purchase table (Y/N): ')
                            if confirm == 'Y' or confirm == 'y': 
                                sql_commands_10 = """DELETE FROM inventory.purchase
                                                WHERE po_number = %s and inventory_id = %s;"""
                                values = (purchase_1,purchase_2)   

                                # Execute SQL to delete purchase
                                execute_sql(sql_commands_10, values)
                                print('Successfully delete data to the database.')
                                saveactivity_log(user,action,action_detail)

                                print('======================================================================================================================================================')
                                print('\033[1m============================================================ product code delete =====================================================================\033[0m')
                                print('======================================================================================================================================================\n\n')
                                read_sql(sql_commands_check, inventory_db)
                                
######################################################################  Delete Sale  ########################################################################
                                
                        elif update_int1 == 'sale' or update_int1 =='2' :
                            action_detail = 'Delete Sale Stock'
                            while True:
                                sale_1 = input('Enter SO number: ')
                                sql_commands_check = '''select * from sale where so_number = '%s';''' %sale_1
                                df_check = read_sql(sql_commands_check, inventory_db)
                                if sale_1.isdigit() and len(df_check) > 0:
                                    break
                                else:
                                    print("Invalid input. Please enter a numeric for SO number.")  
                            
                            sale_2 = input('Enter inventory id: ')
                            print('======================================================================================================================================================')
                            print('\033[1m==================================================== product code selection to delete  ===============================================================\033[0m')
                            print('======================================================================================================================================================\n\n')

                            sql_commands_check = '''select * from sale where so_number = '%s' and inventory_id = '%s';''' %(sale_1,sale_2)
                            read_sql(sql_commands_check, inventory_db)

                            confirm =  input('Do you want to delete (Y/N): ')
                            if confirm == 'Y' or confirm == 'y': 
                                sql_commands_11 = """DELETE FROM inventory.sale
                                                WHERE so_number = %s and inventory_id = %s;"""
                                values = (sale_1,sale_2) 

                                # Execute SQL to delete sale
                                execute_sql(sql_commands_11, values)
                                print('Successfully delete data to the database.')
                                saveactivity_log(user,action,action_detail)
                            
                                print('======================================================================================================================================================')
                                print('\033[1m============================================================== product code delete ===================================================================\033[0m')
                                print('======================================================================================================================================================\n\n')

                                read_sql(sql_commands_check, inventory_db)

###########################################################  option 3.report summary ###################################################################

            elif choice == '3':
                action = 'report summary'
                action_detail = 'Summary all'
                saveactivity_log(user,action,action_detail)
                
                print('=====================================================================================================================================================')
                print('\033[1m====================================================== Top/Bottom Vender by purchase cost ===========================================================\033[0m')
                print('=====================================================================================================================================================\n\n')

                
                sql_commands_12 = '''select vendor_name,sum(dollars) as sum_dollars
                                    from purchase
                                    left join inventory
                                    on purchase.inventory_id = inventory.inventory_id
                                    where vendor_name is not null
                                    group by vendor_name
                                    order by sum(dollars) desc
                                    limit 10'''
                
                sql_commands_13 = '''select vendor_name,sum(dollars) as sum_dollars
                                    from purchase
                                    left join inventory
                                    on purchase.inventory_id = inventory.inventory_id
                                    where vendor_name is not null
                                    group by vendor_name
                                    order by sum(dollars) desc
                                    limit 10,1000'''
                
                df12 = read_sql2(sql_commands_12, inventory_db)
                df13 = read_sql2(sql_commands_13, inventory_db)
                
                fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 8))  # Create subplots 2x1

                # Bar chart
                ax1.bar(df12['vendor_name'], df12['sum_dollars'])
                ax1.set_title('Top 10 vendors by purchase cost')
                ax1.set_xlabel('Vendor Name')
                ax1.set_ylabel('Total Purchase Amount ($)')
                ax1.tick_params(axis='x', rotation=90)
               
                # Horizontal Bar chart
                ax2.barh(df13['vendor_name'], df13['sum_dollars'])
                ax2.set_title('Distribution of Purchase cost among top vendors')
                ax2.set_xlabel('Total Purchase Amount ($)')
                ax2.set_ylabel('Vendor Name')

                plt.tight_layout()
                plt.show()                
                
                print('=====================================================================================================================================================')
                print('\033[1m================================================================ product summary trend ==============================================================\033[0m')
                print('=====================================================================================================================================================\n\n')
 
                sql_commands_14 = '''select item_name,min(sales_qty) as min_sales_qty, max(sales_qty) as max_sales_qty, round(avg(sales_qty),2) as avg_sales_qty
                                from sale
                                left join inventory on sale.inventory_id = inventory.inventory_id
                                where item_name is not null
                                group by 1
                                order by round(avg(sales_qty),2) desc
                                limit 20'''
                df14 = read_sql2(sql_commands_14, inventory_db)
        
                sql_commands_15 = '''select item_name,min(sales_dollar) as min_sales, max(sales_dollar) as max_sales, round(avg(sales_dollar),2) as avg_sales
                                from sale
                                left join inventory on sale.inventory_id = inventory.inventory_id
                                where item_name is not null
                                group by 1
                                order by round(avg(sales_dollar),2) desc
                                limit 20'''
                df15 = read_sql2(sql_commands_15, inventory_db)
                
                sql_commands_16 = '''select item_name,min(quantity) as min_purchase_qty, max(quantity) as max_purchase_qty, round(avg(quantity),2) as avg_purchase_qty
                                from purchase
                                left join inventory on purchase.inventory_id = inventory.inventory_id
                                where item_name is not null
                                group by 1
                                order by round(avg(quantity),2) desc
                                limit 20'''
                df16 = read_sql2(sql_commands_16, inventory_db)
        
                sql_commands_17 = '''select item_name,min(dollars) as min_purchase, max(dollars) as max_purchase, round(avg(dollars),2) as avg_purchase
                                from purchase
                                left join inventory on purchase.inventory_id = inventory.inventory_id
                                where item_name is not null
                                group by 1
                                order by round(avg(dollars),2) desc
                                limit 20'''
                df17 = read_sql2(sql_commands_17, inventory_db)
                

                fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize=(15, 13))
                plt.subplots_adjust(hspace=1)

                data_frames = [df14, df15, df16, df17]
                titles = ['Sale Quantity Summary', 'Sale amount($) Summary', 'Purchase Quantity Summary', 'Purchase amount($) Summary']
                y_labels = ['Average Sale quantity', 'Average Sale amount($)', 'Average Purchase quantity', 'Average Purchase amount($)']

                for i, ax in enumerate([ax1, ax2, ax3, ax4]):
                    df = data_frames[i]
                    ax.bar(df['item_name'], df.iloc[:, 3])
                    ax.set_title(titles[i],fontsize = 12)
                    ax.set_xlabel('Item Name', fontsize=12)
                    ax.set_ylabel(y_labels[i], fontsize=12)
                    ax.tick_params(axis='x', rotation=90, labelsize=12)
                    ax.tick_params(axis='y', labelsize=12)

                plt.tight_layout()
                plt.show()

                print('=====================================================================================================================================================')
                print('\033[1m============================================================= Average lead time =====================================================================\033[0m')
                print('=====================================================================================================================================================\n\n')


                sql_commands_18 = '''select item_id, item_name, min(leadtime)as min_leadtime, max(leadtime)as max_leadtime,avg(leadtime)as avg_leadtime
                                    from (
                                            select purchase.inventory_id,item_id,item_name,
                                            vendor_number,vendor_name,po_date,receiving_date, DATEDIFF(receiving_date, po_date) AS leadtime
                                            from purchase
                                            left join inventory on purchase.inventory_id = inventory.inventory_id) t1
                                    group by item_id,item_name
                                    having avg(leadtime) is not null
                                    order by avg(leadtime) desc'''
                
                df18 = read_sql(sql_commands_18, inventory_db)
                
                sql_commands_19 = '''SELECT
                                t1.vendor_number,
                                t1.vendor_name,
                                AVG(t1.leadtime) AS avg_leadtime,
                                COUNT(t1.item_id) AS purchase_time
                            FROM (
                                SELECT
                                    purchase.inventory_id,
                                    inventory.item_id,
                                    inventory.vendor_number,
                                    inventory.vendor_name,
                                    purchase.po_date,
                                    purchase.receiving_date,
                                    DATEDIFF(purchase.receiving_date, purchase.po_date) AS leadtime
                                FROM purchase
                                LEFT JOIN inventory
                                ON purchase.inventory_id = inventory.inventory_id
                                ) t1 
                            GROUP BY
                                t1.vendor_number,
                                t1.vendor_name
                            ORDER BY
                                avg_leadtime DESC'''
                df19 = read_sql(sql_commands_19, inventory_db)
                
                # Create a bar graph to compare AvgLT with purchase_time
                fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 6)) 
                fig.suptitle("Lead Time by Item Distribution", color='tomato', weight='bold', fontsize=12)

                # Subplot 1: Histogram of average lead time
                ax1.hist(df18["avg_leadtime"], bins=200, density=False, edgecolor="k", label="Average Lead Time")
                ax1.set_xlabel("Average Lead Times (Days)")
                ax1.set_ylabel("Frequency")
                ax1.legend()

                # Subplot 2: Bar chart of vendor's average lead time and line plot of purchase frequency
                color = 'tab:blue'
                ax2.bar(df19.index, df19['avg_leadtime'], color=color, alpha=0.6)
                ax2.set_xlabel('Vendor')
                ax2.set_ylabel('Average Lead Time (Days)', color=color)
                ax2.tick_params(axis='y', labelcolor=color)

                # Creating a second y-axis to plot purchase_time on the same subplot (ax2)
                ax3 = ax2.twinx()
                color = 'tab:red'
                ax3.set_ylabel('Frequency of Purchase (Times)', color=color)
                ax3.plot(df19.index, df19['purchase_time'], color=color, marker='o', linestyle='None')
                ax3.tick_params(axis='y', labelcolor=color)

                plt.show()
                
                sql_commands_20 = '''WITH table1 as (
                                            SELECT
                                                t1.vendor_number,
                                                t1.vendor_name,
                                                AVG(t1.leadtime) AS avg_leadtime,
                                                COUNT(t1.item_id) AS purchase_time
                                            FROM (
                                                SELECT
                                                    purchase.inventory_id,
                                                    inventory.item_id,
                                                    inventory.vendor_number,
                                                    inventory.vendor_name,
                                                    purchase.po_date,
                                                    purchase.receiving_date,
                                                    DATEDIFF(purchase.receiving_date, purchase.po_date) AS leadtime
                                                FROM purchase
                                                LEFT JOIN inventory
                                                ON purchase.inventory_id = inventory.inventory_id
                                            ) t1
                                            GROUP BY
                                                t1.vendor_number,
                                                t1.vendor_name
                                            ORDER BY
                                                avg_leadtime DESC),

                                Stats AS (
                                    SELECT
                                        AVG(avg_leadtime) AS avg_AvgLT,
                                        AVG(purchase_time) AS avg_purchase_time,
                                        COUNT(*) AS n
                                        FROM table1),

                                Covariance AS (
                                    select covariance_nom/(n - 1) as covariance
                                    from (
                                        SELECT
                                            SUM((avg_leadtime - avg_AvgLT)*(purchase_time - avg_purchase_time)) AS covariance_nom
                                        FROM table1, Stats) t2, Stats),

                                Variance_AvgLT AS (
                                    select var_AvgLT_nom/ (n - 1) as var_AvgLT
                                    from (
                                        SELECT
                                            (SUM(POWER(avg_leadtime - avg_AvgLT, 2)) ) AS var_AvgLT_nom
                                        FROM table1, Stats) t3, Stats),

                                Variance_purchase_time AS (
                                    select var_purchase_time_nom/(n - 1) as var_purchase_time
                                    from (
                                        SELECT
                                            (SUM(POWER(purchase_time - avg_purchase_time, 2))) AS var_purchase_time_nom
                                        FROM table1, Stats) t4, Stats)

                                SELECT
                                    round(covariance / SQRT(var_AvgLT * var_purchase_time),5) AS correlation_coefficient
                                FROM Covariance, Variance_AvgLT, Variance_purchase_time;'''

                print('\033[1m============================================================== Correlation_Coefficient ==============================================================\033[0m')
                df20 = read_sql(sql_commands_20, inventory_db)                
                
                print('=====================================================================================================================================================')
                print('\033[1m================================================================ Alerting out of stock ==============================================================\033[0m')
                print('=====================================================================================================================================================\n\n')

                print('\033[1mCriteria based on balance stock < 30% of average sale per month and average sale per month >= 20 bottles\033[0m\n')

                sql_commands_21 = """select city, item_name, size,onhand as onhand_qty,
                                    COALESCE(purchase_quantity,0) as purchase_qty,
                                    COALESCE(sales_qty,0) as sales_qty, sales_month_avg,
                                    COALESCE(balance_stock,0) as balance_stock
                                    from 
                                        (select city,b.inventory_id,b.onhand,b.cost_per_unit,b.amount,
                                        t1.purchase_quantity, t1.dollars,
                                        t2.sales_qty, item_name, size, t2.sales_month_avg,
                                        b.onhand+COALESCE(t1.purchase_quantity,0)-COALESCE(t2.sales_qty,0) as balance_stock,
                                        b.onhand+COALESCE(t1.purchase_quantity, 0) as onhand_purchase,
                                        b.amount+t1.dollars as total_cost_onhand_purchase,
                                        round((b.amount+COALESCE(t1.dollars,0))/(b.onhand+COALESCE(t1.purchase_quantity, 0)),2) as avg_cost,
                                        round(((b.onhand+COALESCE(t1.purchase_quantity,0)-COALESCE(t2.sales_qty,0)) * (b.amount+COALESCE(t1.dollars,0))/(onhand+COALESCE(purchase_quantity, 0))),2) as inventory_value$
                                        from begin as b
                                        left join (select inventory_id, COALESCE(sum(dollars),0) as dollars, COALESCE(sum(quantity),0) as purchase_quantity 
                                        from purchase where receiving_date between '2016-01-01' and '2016-02-28' group by inventory_id) as t1
                                        on b.inventory_id = t1.inventory_id
                                            
                                        left join (select inventory_id, COALESCE(sum(sales_qty),0) as sales_qty, 
                                        round(SUM(sales_qty) / (select max(extract(month from sales_date)) from sale),2) AS sales_month_avg
                                        from sale where sales_date between '2016-01-01' and '2016-02-28' group by inventory_id) as t2
                                        on b.inventory_id = t2.inventory_id
                                            
                                        left join inventory 
                                        on b.inventory_id = inventory.inventory_id ) t3
    
                                        where balance_stock < 3*sales_month_avg/10 and sales_month_avg >= 20
                                        order by sales_month_avg desc, balance_stock desc;"""
                
                df21 = read_sql(sql_commands_21, inventory_db)
            
                print('=====================================================================================================================================================')
                print('\033[1m================================================================= Top item margin ===================================================================\033[0m')
                print('=====================================================================================================================================================\n\n')
                
                print('\033[1mจากข้อมูลข้างล่างแสดง ค่า margin % ของ productต่อชิ้น เทียบกับ กำไรรวม เพื่อนำไปจัดวางแคมเปนในเพิ่มยอดขายของ\nให้กับproduct ที่มี % marginแต่ยอดขายไม่สูงมากเพื่อเป็นการเพิ่ม total profit ให้กับบริษัทได้\033[0m\n')

                
                sql_commands_22_1 = '''select item_name, round(coalesce(avg(sales_price),0),2) as `sale_price($)`,
                                    round(avg(purchase_price),2) as `purchase_price($)`,
                                    round(coalesce(avg(sales_price),0),2) - round(avg(purchase_price),2) as `margin($)`,
                                    round((coalesce(avg(sales_price),0)-avg(purchase_price))*100/avg(purchase_price),2) as `margin(%)`,
                                    sum(sales_qty) as sales_qty,
                                    round(sum(sales_dollar),2) as `tot_sale($)`,
                                    round((sum(sales_qty) * (round(coalesce(avg(sales_price),0),2) - round(avg(purchase_price),2))),2) as `profit($)`
                                    
                                    from purchase
                                    left join sale on purchase.inventory_id = sale.inventory_id
                                    left join inventory on purchase.inventory_id = inventory.inventory_id
                                    where sales_qty > 0
                                    group by item_name
                                    order by round(coalesce(avg(sales_price),0),2) - round(avg(purchase_price),2)   desc
                                    limit 10;'''
                
                df22 = read_sql(sql_commands_22_1, inventory_db)
                
                print('\n\033[1mถ้าเทียบจาก tableข้างล่างที่เป็น mass product มี total profit จากจำนวนที่ขายได้มากแต่มี % margin ไม่มากนัก เราก็ไม่ต้องเน้น แคมเปนกับ puduct เหล่านี้มากนัก\033[0m\n')
                
                sql_commands_22_2 = '''select item_name, round(coalesce(avg(sales_price),0),2) as `sale_price($)`,
                                    round(avg(purchase_price),2) as `purchase_price($)`,
                                    round(coalesce(avg(sales_price),0),2) - round(avg(purchase_price),2) as `margin($)`,
                                    round((coalesce(avg(sales_price),0)-avg(purchase_price))*100/avg(purchase_price),2) as `margin(%)`,
                                    sum(sales_qty) as sales_qty,
                                    round(sum(sales_dollar),2) as `tot_sale($)`,
                                    round((sum(sales_qty) * (round(coalesce(avg(sales_price),0),2) - round(avg(purchase_price),2))),2) as `profit($)`
                                    
                                    from purchase
                                    left join sale on purchase.inventory_id = sale.inventory_id
                                    left join inventory on purchase.inventory_id = inventory.inventory_id
                                    where sales_qty > 0
                                    group by item_name
                                    order by sum(sales_qty) desc
                                    limit 10;'''
                
                df22 = read_sql(sql_commands_22_2, inventory_db)
                
                
                sql_commands_23 = '''select item_name, round(coalesce(avg(sales_price),0),2) - round(avg(purchase_price),2) as `margin($)`,
                                    round(avg(purchase_price),2) as `purchase_price($)`,
                                    round(coalesce(avg(sales_price),0),2) as `sale_price($)`
                                    
                                    from purchase
                                    left join sale on purchase.inventory_id = sale.inventory_id
                                    left join inventory on purchase.inventory_id = inventory.inventory_id
                                    where sales_qty > 0
                                    group by item_name
                                    order by round(coalesce(avg(sales_price),0),2) - round(avg(purchase_price),2)  desc
                                    limit 10;'''
                
                df23 = read_sql2(sql_commands_23, inventory_db)
                
                sql_commands_24 = '''select item_name, round(coalesce(avg(sales_price),0),2) - round(avg(purchase_price),2) as `margin($)`,
                                    round(avg(purchase_price),2) as `purchase_price($)`,
                                    round(coalesce(avg(sales_price),0),2) as `sale_price($)`
                                    
                                    from purchase
                                    left join sale on purchase.inventory_id = sale.inventory_id
                                    left join inventory on purchase.inventory_id = inventory.inventory_id
                                    where sales_qty > 0
                                    group by item_name
                                    order by sum(sales_qty) desc
                                    limit 10;'''
                
                df24 = read_sql2(sql_commands_24, inventory_db)     
                
                #df23.set_index('item_name').plot(kind='bar', stacked=True, figsize=(10, 6), width=0.8)

                #plt.xlabel('Item Name', fontweight='bold')
                #plt.ylabel('Price($)', fontweight='bold')
                #plt.title('Top item margin')

                # แสดงแผนภูมิ
                #plt.legend(title='Categories', bbox_to_anchor=(1.05, 1), loc='upper left')

                #plt.show()

                fig, axs = plt.subplots(1, 2, figsize=(15, 6))

                # Plot ข้อมูลใน DataFrame df23 ใน subplot ซ้าย
                df23.set_index('item_name').plot(kind='bar', stacked=True, ax=axs[0], width=0.8)
                axs[0].set_xlabel('Item Name', fontweight='bold')
                axs[0].set_ylabel('Price($)', fontweight='bold')
                axs[0].set_title('Top item margin')
                axs[0].legend(title='Categories', bbox_to_anchor=(1.05, 1), loc='upper left')

                # Plot ข้อมูลใน DataFrame df24 ใน subplot ขวา
                df24.set_index('item_name').plot(kind='bar', stacked=True, ax=axs[1], width=0.8)
                axs[1].set_xlabel('Item Name', fontweight='bold')
                axs[1].set_ylabel('Price($)', fontweight='bold')
                axs[1].set_title('Top item sales quantity')
                axs[1].legend(title='Categories', bbox_to_anchor=(1.05, 1), loc='upper left')

                plt.tight_layout()  # จัดเรียง subplot ให้พอดี
                plt.show()                
                
                
                
                print('=====================================================================================================================================================')
                print('\033[1m==================================================================== Export data ===================================================================\033[0m')
                print('=====================================================================================================================================================\n\n')
                
                export = input('Do you need to export? (Y/N): ')
                if export == 'Y' or export == 'y':
                    action = 'Export data'
                    action_detail = 'Export all data'
                    saveactivity_log(user,action,action_detail)

                    df12.to_csv('D:/DSDS4002/Project_final/output/Top Vender by purchase cost.csv',index=False,mode='w')
                    df13.to_csv('D:/DSDS4002/Project_final/output/Bottom Vender by purchase cost.csv',index=False,mode='w')
                    df14.to_csv('D:/DSDS4002/Project_final/output/product summary trend_v1.csv',index=False,mode='w')
                    df15.to_csv('D:/DSDS4002/Project_final/output/product summary trend_v2.csv',index=False,mode='w')
                    df16.to_csv('D:/DSDS4002/Project_final/output/product summary trend_v3.csv',index=False,mode='w')
                    df17.to_csv('D:/DSDS4002/Project_final/output/product summary trend_v4.csv',index=False,mode='w')
                    df18.to_csv('D:/DSDS4002/Project_final/output/Average lead time.csv',index=False,mode='w')                    
                    df21.to_csv('D:/DSDS4002/Project_final/output/Alerting out of stock.csv',index=False,mode='w') 
                    df22.to_csv('D:/DSDS4002/Project_final/output/Top item margin.csv',index=False,mode='w')
                
                    print('Successful export')
                    print('Output data are under D:/DSDS4002/Project_final/output/\n')
                    print('file name:Top Vender by purchase cost.csv')
                    print('file name:Bottom Vender by purchase cost.csv')
                    print('file name:product summary trend_v1.csv')
                    print('file name:product summary trend_v2.csv')
                    print('file name:product summary trend_v3.csv')
                    print('file name:product summary trend_v4.csv')
                    print('file name:Average lead time.csv')
                    print('file name:Alerting out of stock.csv')
                    print('file name:Top item margin.csv')

                    
############################################################################  end program ########################################################################            
                            
except Exception as e:
    print('Something went wrong. Please try again')
    if inventory_db is not None: inventory_db.rollback()
    print( f'>>> Error {e}' )

finally:
    if inventory_db is not None: inventory_db.close()
    print('\nDatabase closed.')
    action = 'Log out'
    action_detail = 'user log out'
    saveactivity_log(user,action,action_detail)
```
